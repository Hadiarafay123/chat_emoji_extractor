```python
from google.colab import files

# This will open a file picker for image upload
uploaded = files.upload()

# Check uploaded file
for filename in uploaded.keys():
    print(f'Uploaded file: {filename}')
from PIL import Image
import matplotlib.pyplot as plt

img = Image.open(next(iter(uploaded)))
plt.imshow(img)
plt.axis('off')
plt.title("Uploaded Chat Screenshot")
plt.show()

!pip install emoji

import emoji
import re
import json

# Sample chat lines
lines = [
    "Heyyy bestieee!! ",
    "Guess what??? I just saw the cutest cat video everrr ",
    "Also, Im craving bubble tea again like always ",
    "Are you free this weekend? Lets do something funnn ",
    "I found this aesthetic caf we have to try ",
    "Alsoooo I need a rant session asap haha ",
    "Okay Ill stop spamming now... maybe ",
    "Miss youuuu! Text me backkkk! "
]

timestamp = "2025-05-24T16:19:00Z"

# Extract emojis and clean text
messages = []
for line in lines:
    emojis = [ch for ch in line if ch in emoji.EMOJI_DATA]
    text = ''.join(ch for ch in line if ch not in emoji.EMOJI_DATA).strip()
    messages.append({
        "timestamp": timestamp,
        "text": text,
        "emojis": emojis
    })

# Format as JSON
chat_json = {
    "messages": messages
}

# Print formatted output
print(json.dumps(chat_json, indent=2, ensure_ascii=False))


```
